```python

top_3_countries = (
    df.assign(country=df['country'].str.split(', '))
      .explode('country')
      .groupby(['type', 'country'])
      .size()
      .groupby(level=0)
      .nlargest(3)
)

print(top_3_countries)
```

#### **Q77.** Compute a rolling count of titles added per month over time after sorting the DataFrame by a proper datetime index.

```python
df['date_added'] = pd.to_datetime(df['date_added'])

monthly_titles = (
    df.sort_values('date_added')
      .set_index('date_added')
      .resample('ME')
      .size()
)

rolling_count = monthly_titles.rolling(window=3).sum()

rolling_count
```
